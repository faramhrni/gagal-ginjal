```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.model_selection import KFold, train_test_split, cross_val_predict, cross_val_score
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, classification_report
from sklearn.preprocessing import StandardScaler
from sklearn import svm, datasets
import plotly.express as px
from sklearn .ensemble import RandomForestClassifier
from sklearn .ensemble import VotingClassifier , BaggingClassifier , StackingClassifier
```

```python
df=pd.read_csv('new_model.csv')
df.head()
```

```python
x=df.drop(columns='Class')
y=df['Class']
scaler=StandardScaler()
scaler.fit(x)
standar=scaler.transform(x)
```

```python
x=standar
y=df['Class']
```

# **PERCANTAGE SPLIT**

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.4, random_state = 50)
```

```python
SVM = svm.SVC(kernel = 'poly', degree=2, C=10, gamma=10)
SVM.fit(x_train,  y_train)
predictionSVM = SVM.predict(x_train)
trainingSVM = accuracy_score(predictionSVM ,y_train)
y_testSVM = SVM.predict(x_test)
test_accuracySVM = accuracy_score(y_test,y_testSVM)

ALR = LogisticRegression()
ALR.fit(x_train,  y_train)
predictionALR = ALR.predict(x_train)
trainingALR = accuracy_score(predictionALR ,y_train)
y_testALR = ALR.predict(x_test)
test_accuracyALR = accuracy_score(y_test,y_testALR)

C45 = DecisionTreeClassifier(random_state = 1)
C45.fit(x_train,  y_train)
predictionC45 = C45.predict(x_train)
trainingC45 = accuracy_score(predictionC45 ,y_train)
y_testC45 = C45.predict(x_test)
test_accuracyC45 = accuracy_score(y_test,y_testC45)

ENSEMBLE = StackingClassifier(estimators=[('SVM', SVM), ('ALR', ALR), ('C45', C45)])
ENSEMBLE.fit(x_train, y_train)
predictionENSEMBLE = ENSEMBLE.predict(x_train)
trainingENSEMBLE = accuracy_score(predictionENSEMBLE ,y_train)
y_testENSEMBLE = ENSEMBLE.predict(x_test)
test_accuracyENSEMBLE = accuracy_score(y_test,y_testENSEMBLE)
```

```python
conf_matrixSVM = confusion_matrix(y_test, y_testSVM)
print("Confusion Matrix SVM:")
print(conf_matrixSVM)

conf_matrixALR = confusion_matrix(y_test, y_testALR)
print("Confusion Matrix ALR:")
print(conf_matrixALR)

conf_matrixC45 = confusion_matrix(y_test, y_testC45)
print("Confusion Matrix C4.5:")
print(conf_matrixC45)

conf_matrixENSEMBLE = confusion_matrix(y_test, y_testENSEMBLE)
print("Confusion Matrix ENSEMBLE:")
print(conf_matrixENSEMBLE)

```

```python
from sklearn.metrics import classification_report

print("\nclassification_report SVM:")
print(classification_report(y_test, y_testSVM))
print("\nclassification_report ALR:")
print(classification_report(y_test, y_testALR))
print("\nclassification_report C4.5:")
print(classification_report(y_test, y_testC45))
print("\nclassification_report ENSMBLE:")
print(classification_report(y_test, y_testENSEMBLE))

#print("Akurasi Data Training SVM: {:.2f}%". format(trainingSVM*100))
#print("Akurasi Data Training ALR: {:.2f}%". format(trainingALR*100))
#print("Akurasi Data Training C4.5: {:.2f}%". format(trainingC45*100))
#print("Akurasi Data Training Ensemble: {:.2f}%". format(trainingENSEMBLE*100))
#print("\n")
print("Akurasi Data Testing SVM: {:.2f}%". format(test_accuracySVM*100))
print("Akurasi Data Testing ALR: {:.2f}%". format(test_accuracyALR*100))
print("Akurasi Data Testing C4.5: {:.2f}%". format(test_accuracyC45*100))
print("Akurasi Data Testing Ensemble: {:.2f}%". format(test_accuracyENSEMBLE*100))
```

# **K-FOLD Cross Validassion**

```python
kfold = KFold(n_splits=4, shuffle=True, random_state=55)
acc_scores = []
for train_index, test_index in kfold.split(x, y):
  x_train, x_test = x[train_index], x[test_index]
  y_train, y_test = y.iloc[train_index], y.iloc[test_index]

  SVM = svm.SVC(kernel = 'poly', degree=2, C=10, gamma=10)
  SVM.fit(x_train,  y_train)
  predictionSVM = SVM.predict(x_train)
  trainingSVM = accuracy_score(predictionSVM ,y_train)
  y_testSVM = SVM.predict(x_test)
  test_accuracySVM = accuracy_score(y_test,y_testSVM)

  ALR = LogisticRegression()
  ALR.fit(x_train,  y_train)
  predictionALR = ALR.predict(x_train)
  trainingALR = accuracy_score(predictionALR ,y_train)
  y_testALR = ALR.predict(x_test)
  test_accuracyALR = accuracy_score(y_test,y_testALR)

  C45 = DecisionTreeClassifier(random_state = 1)
  C45.fit(x_train,  y_train)
  predictionC45 = C45.predict(x_train)
  trainingC45 = accuracy_score(predictionC45 ,y_train)
  y_testC45 = C45.predict(x_test)
  test_accuracyC45 = accuracy_score(y_test,y_testC45)

  ENSEMBLE = StackingClassifier(estimators=[('SVM', SVM), ('ALR', ALR), ('C45', C45)])
  ENSEMBLE.fit(x_train, y_train)
  predictionENSEMBLE = ENSEMBLE.predict(x_train)
  trainingENSEMBLE = accuracy_score(predictionENSEMBLE ,y_train)
  y_testENSEMBLE = ENSEMBLE.predict(x_test)
  test_accuracyENSEMBLE = accuracy_score(y_test,y_testENSEMBLE)
```

```python
conf_matrixSVM = confusion_matrix(y_test, y_testSVM)
print("Confusion Matrix SVM:")
print(conf_matrixSVM)

conf_matrixALR = confusion_matrix(y_test, y_testALR)
print("Confusion Matrix ALR:")
print(conf_matrixALR)

conf_matrixC45 = confusion_matrix(y_test, y_testC45)
print("Confusion Matrix C4.5:")
print(conf_matrixC45)

conf_matrixENSEMBLE = confusion_matrix(y_test, y_testENSEMBLE)
print("Confusion Matrix ENSEMBLE:")
print(conf_matrixENSEMBLE)
```

```python
print("\nclassification_report SVM:")
print(classification_report(y_test, y_testSVM))
print("\nclassification_report ALR:")
print(classification_report(y_test, y_testALR))
print("\nclassification_report C4.5:")
print(classification_report(y_test, y_testC45))
print("\nclassification_report ENSEMBLE:")
print(classification_report(y_test, y_testENSEMBLE))

#print("Akurasi Data Training SVM: {:.2f}%". format(trainingSVM*100))
#print("Akurasi Data Training ALR: {:.2f}%". format(trainingALR*100))
#print("Akurasi Data Training C4.5: {:.2f}%". format(trainingC45*100))
#print("Akurasi Data Training Ensemble: {:.2f}%". format(trainingENSEMBLE*100))
print("\n")
print("Akurasi Data Testing SVM: {:.2f}%". format(test_accuracySVM*100))
print("Akurasi Data Testing ALR: {:.2f}%". format(test_accuracyALR*100))
print("Akurasi Data Testing C4.5: {:.2f}%". format(test_accuracyC45*100))
print("Akurasi Data Testing Ensemble: {:.2f}%". format(test_accuracyENSEMBLE*100))
```
